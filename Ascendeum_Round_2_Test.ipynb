```python
# Direct approach
bidder_prebid_count = df.groupby('bidder')['prebid_win_count'].sum()
bidder_predi_total = df.groupby('bidder')['response_count'].sum()

win_rate = (bidder_prebid_count/bidder_predi_total)*100

win_rate
```

```python
bidder_stats = df.groupby('bidder').agg(total_response = ("response_count", "sum"),
                                        total_wins = ("prebid_win_count", "sum"))
```

```python
bidder_stats
```

```python
bidder_stats['win_rate'] = (bidder_stats['total_wins'] / bidder_stats['total_response'])*100
bidder_stats
```

```python
## Final Output as desired with column names and their count
bidder_stats_sorted = bidder_stats.sort_values(by="win_rate", ascending=False)
bidder_stats_sorted
```
